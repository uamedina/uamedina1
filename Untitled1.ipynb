edad y altura de los arboles del bosque


```r
#vectores de datos con la edad en años y la altura de los arboles en metros
edad_arboles <- c (5, 10, 15, 20, 25, 30, 35, 40)
altura_arboles <- c (2.5, 4.8, 7.2, 9.5, 12.1, 12.5, 14, 13.8)

arboles <- data.frame(edad_arboles, altura_arboles)
arboles

#funcion summary
summary(arboles)
#calculo del coeficiente de correlacion con la funcion cor
correlacion <- cor(arboles$edad_arboles, arboles$altura_arboles)
correlacion
```

si es un numero cercano a 1: correlacion lineal positiva
si es cercano a 0: no hay correlacion
si es cercano a -1: la correlacion es negativa



```r
plot(arboles$edad_arboles, arboles$altura_arboles, xlab = "Edad (años)", ylab = "Altura (metros)", main = "Relación entre Edad y Altura de Árboles")
```

ejercicio de correlacion en ciencias de la tierra

mediciones en puntos geograficos de las siguientes variables:
1. ph
2. contenido en materia organica
3. temperatura
4. humedad relativa
5. precipitacion


```r
ph <- c (5.8, 6.5, 7.2, 5.9)
materia <- c (3.1, 4.5, 5, 3.8)
temperatura <- c (22, 25, 30, 21)
humedad <- c (70, 65, 55, 80)
precipitacion <-  c (50, 80, 140, 120)

# creacion de marco de datos
suelos <- data.frame (ph, materia, temperatura, humedad, precipitacion)
suelos

# analisis de correlacion multiple (correlacion de pearson)
correlation <- cor (suelos, method ="pearson")
correlation
#correlacion de spearman
correlation2 <- cor (suelos, method = "spearman")
correlation2
#correlacion de kendall
correlation3 <- cor (suelos, method ="kendall")
correlation3

boxplot (suelos)
plot (suelos)
```
